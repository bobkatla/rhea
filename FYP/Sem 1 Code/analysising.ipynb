```python
df['EV'] = 0
df.loc[df['cluster'] == 8, 'EV'] = 1
df
```

```python
X = df[['INC', 'HHSIZE', 'TOTALVEHS', 'own', 'DWELLTYPE_Separate House', 'age_profile']]
Y = df['EV']

X = sm.add_constant(X)

model = sm.Logit(Y, X)
results = model.fit()
print(results.summary())
```

```python
results_summary = results.summary()
summary_df = pd.read_html(results_summary.tables[1].as_html(), header=0, index_col=0)[0]
summary_df.to_excel('logistic_regression_summary.xlsx')
```
