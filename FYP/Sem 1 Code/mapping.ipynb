```python
import geopandas as gpd
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
shapefile_folder = '1270055001_sa2_2016_aust_shape\SA2_2016_AUST.shp'
australia = gpd.read_file(shapefile_folder)
df_household_import = pd.read_csv("synthetic_households.csv")
```

```python
df_household = df_household_import.copy()
```

```python
SA4 = ('203', '213', '210', '210', '209', '214', '206', '207', '208', '211', '212')
select = pd.DataFrame({'SA2': df_household[df_household['SA2'].astype(str).str.startswith(SA4)]['SA2'].unique()}).astype(str)
combined_df = australia.merge(select, left_on='SA2_MAIN16', right_on='SA2', how='inner')
combined_df.plot()
```
