```python
import pandas as pd 
```

```python
df = pd.read_csv("Data\syn_2021_HH.csv")
```

```python
df = df[~(df["dwelltype"] == "Missing")]

column_mapping = {
    'totalvehs': 'TOTALVEHS',
    'hhsize': 'HHSIZE',
    'hhinc': 'INC'
}
df = df.rename(columns=column_mapping)

new_column_order = ['SA1', 'INC', 'TOTALVEHS', 'HHSIZE', 'dwelltype', 'owndwell']
df = df[new_column_order]
df.rename(columns={'SA1': 'SA1_CODE21'}, inplace=True)
df["SA1_CODE21"] = df["SA1_CODE21"].astype(str)
```

## One Hot Encoding

```python
df_onehot = df.copy()
df_onehot = pd.get_dummies(df_onehot, columns=['dwelltype', 'owndwell'])
df_onehot
```

```python
df_onehot.to_csv('Data\clean_data.csv', index=False)
```
