```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate random data for clustering with greater separation
np.random.seed(0)

# Create three separate clusters
cluster1 = np.random.rand(5, 2) * 2
cluster2 = np.random.rand(5, 2) * 2 + np.array([3, 3])
cluster3 = np.random.rand(5, 2) * 2 + np.array([0, 5])

# Combine the clusters into a single dataset
data = np.vstack((cluster1, cluster2, cluster3))

# Perform K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
cluster_labels = kmeans.fit_predict(data)

# Plot the data points and cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='viridis', s=1000)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200)

# Draw lines between cluster centers
for i in range(len(kmeans.cluster_centers_)):
    for j in range(i+1, len(kmeans.cluster_centers_)):
        plt.plot([kmeans.cluster_centers_[i][0], kmeans.cluster_centers_[j][0]], 
                 [kmeans.cluster_centers_[i][1], kmeans.cluster_centers_[j][1]], 
                 color='black', linestyle='--', alpha=0.7)

plt.legend()
plt.axis('off')
plt.show()
```
